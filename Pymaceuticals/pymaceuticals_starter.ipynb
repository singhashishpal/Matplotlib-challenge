```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merging study_results and mouse_metadata using 'left join' on the "Mouse ID"
mouse_study_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='left')

# Display the data table for preview
mouse_study_df.head()
```

```python
# Checking the number of mice.
mouse_metadata['Mouse ID'].count()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice_data = mouse_study_df[mouse_study_df.duplicated(subset=['Mouse ID','Timepoint'], keep=False)]
duplicate_mice_data['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID.
optional_data = mouse_study_df[mouse_study_df['Mouse ID'] == 'g989']
optional_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

# Merging both dataframes with indicator
unique_mice_data = pd.merge(mouse_study_df, optional_data, how='left', indicator=True)

# Filter out rows that are in optional_data                         --> # https://stackoverflow.com/questions/39880627/in-pandas-how-to-delete-rows-from-a-data-frame-based-on-another-data-frame
unique_mice_s = unique_mice_data[unique_mice_data['_merge'] == 'left_only'].drop(columns=['_merge'])
unique_mice = pd.DataFrame(unique_mice_s)
unique_mice.head()
```

```python
# Checking the number of mice in the clean DataFrame.
unique_mice['Mouse ID'].nunique()  # Counting the unique number of Mouse ID.
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


# Use groupby for each drug regimen:
mice_grouped = unique_mice.groupby('Drug Regimen')

# Mean of the tumor volume.
tumor_mean = mice_grouped['Tumor Volume (mm3)'].mean()

# Median of the tumor volume.
tumor_median = mice_grouped['Tumor Volume (mm3)'].median()

# Variance of the tumor volume.
tumor_variance = mice_grouped['Tumor Volume (mm3)'].var()

# Standard deviation of the tumor volume.
tumor_std_dev = mice_grouped['Tumor Volume (mm3)'].std()

# SEM of the tumor volume.
tumor_SEM = mice_grouped['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
data = {
    'Mean Tumor Volume' : tumor_mean,
    'Median Tumor Volume' : tumor_median,
    'Tumor Volume Variance' : tumor_variance,
    'Tumor Volume Std. Dev.' : tumor_std_dev,
    'Tumor Volume Std. Err.' : tumor_SEM
}

drug_regimen = pd.DataFrame(data)
drug_regimen
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line                   # https://datagy.io/pandas-groupby/

drug_regimen_agg = unique_mice.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
drug_regimen_agg
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
observed_timepoints = mice_grouped['Timepoint'].count()                     # Count for each of the Drug Regimen after grouping them
timepoints_sorted = observed_timepoints.sort_values(ascending=False)        # Sorting the timepoints in descending order

# Plotting the bar chart using pandas with x and y labels.
timepoints_sorted.plot(kind = 'bar', xlabel = 'Drug Regimen', ylabel = '# of Observed Mouse Timepoints', color= 'r', alpha = 0.5)
plt.legend()
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
drug_regimen_name = unique_mice['Drug Regimen'].unique()                        # Storing the unique name values into an variable

#  Plotting the bar chart using Pyplot (matplotlib).
plt.bar(drug_regimen_name, timepoints_sorted, align="center")
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints", fontsize = 10)
plt.xticks(rotation=90)                                                         # This will rotate the xlabel to 90 degrees
plt.show()
```

```python
# Generate a pie chart, using Pandas, showing the distribution of unique female versus male mice used in the study
mice_sex = unique_mice['Sex']                                                       # Finding out the sex of the mice

# Get the value counts of mice with their gender
unique_mice_sex = mice_sex.value_counts()

# Make the pie chart
labels = ['Male', 'Female']
unique_mice_sex.plot(y = mice_sex, labels = labels, kind = 'pie', autopct= "%1.1f%%")
plt.show()
```
